### 2-1. Line breaking
- 일부 데이터에서 줄바꿈되지 않은 문장이 존재하여 변환

```python
from kss import split_sentences
from konlpy.tag import Mecab

idx= []; org_idx = []
for i in range(len(train)):
    if '\n' not in train['text'][i]:
        idx.append(i)
    else:
        org_idx.append(i)

refine_df = train.iloc[idx, :];org_df = train.iloc[org_idx, :]
for i in idx:
    refine_df.loc[i,'text'] = '\n'.join(split_sentences(refine_df.loc[i,'text'],backend = 'mecab')) 

final_df = pd.concat([org_df,refine_df], axis = 0)
final_df = final_df.sort_index(ascending=True)
```

```python
final_df.to_csv('/content/drive/MyDrive/공모전/final_train.csv', index=False)
```

### 2-2. Preprocessing

```python
import re
import emoji
from unidecode import unidecode
import numpy as np

#remove html tags
def removeHTML(x):
    html=re.compile(r'<.*?>')
    return html.sub(r'',x)
    

def dataPreprocessing(x):
    x = removeHTML(x)
    x = emoji.demojize(x, delimiters=(" ", " "))
    x = re.sub(r'[^ 가-힣A-Za-z0-9\'.,?!/\n\r()%\[\]\-+~$&"]·', '', x)
    x = re.sub(r'[ ]+', ' ', x) # 공백 두칸 이상 지우기
    x = x.strip()
    return x
```

```python
train["text"]=train["text"].apply(lambda x: dataPreprocessing(x))
```
